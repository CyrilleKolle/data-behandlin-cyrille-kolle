## Concat
- concatenates pandas objects along an axis (rows, colums)
- there are set logics


```python
df1, df2
```

```python
pd.concat([df1, df2], axis = 0) #axis=0 default
```

```python
#puts all cols from df1 and then all cols from df2
# those indices that are missing in one df 
pd.concat([df1,df2], axis="columns")
```

```python
df1, df3
```

```python
#concatenates in axis rows, df1 rows first then df2 rows, NANs are used for missing values
pd.concat([df1, df3], axis='rows')
```

```python
# per default outer joint
pd.concat([df1, df3], axis='columns')
```

```python
# intersection between df1 and df2 on the indices
pd.concat([df1, df3], axis='columns', join="inner")
```

## Merge

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K0", "K1", "K2"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
left, right
```

```python
#k3 not included -> k3 from right not included i.e inner joint
pd.merge(left, right, on="key", indicator=True)
```

```python
# syntax with method
left.merge(right, on="key", indicator=True)
```

```python
# note that left does not have a k3
# so in row index 4 onöly merge from right
# outer join -> unioin between left and right
left.merge(right, on="key",how="outer", indicator=True)
```

# new example

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "A": ["A0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
left, right
```

```python
#merge on both key1 and key2 columns, and an inner join - intersection
left.merge(right, on=["key1", "key2"], indicator=True)
```

```python
#A_x left and A_y is right
left.merge(right, on=["key1", "key2"], indicator=True, how="left")
```

```python
left.merge(right, on=["key1", "key2"], indicator=True, how="right")
```

## Join
- uses merge internally
- combines columns

```python
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
)


right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)
left, right
```

```python
# note we join on k1 which only exists in left -> so this is a left join
left.join(right)
```

```python
# indices from right -> right join
left.join(right, how="right")
```
