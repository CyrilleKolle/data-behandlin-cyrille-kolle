```python
import pandas as pd

befolkning_1900 = pd.read_html("https://sv.wikipedia.org/wiki/Sveriges_demografi")[5]
befolkning_1900.head()
```

```python
befolkning_1900 = befolkning_1900.rename(columns={'Unnamed: 0':'Year', 'Medel- folkmängd':'MeanPop', 'Födda':'Births', 'Döda':'Deaths', 'Naturlig befolkningsförändring':'NatPopChange','Nativitet (per 1000)':'Birtrate1000', 'Dödstal (per 1000)':'Deathrate1000', 'Naturlig befolkningsförändring (per 1000)':'NatChange1000', 'Total fertilitet':'TotFertility'})
```

```python
befolkning_1900.head()
```

```python
import seaborn as sns
sns.barplot(data=befolkning_1900, x="Deaths", y="Year")
```

```python
sns.lineplot(data=befolkning_1900, y="Total fertilitet", x="Unnamed: 0")
```

```python
befolkning_1900.head()
```

```python
befolkning_1900.tail()
```

```python
befolkning_1900
```

```python
befolkning_1900["MeanPop"] = befolkning_1900["MeanPop"].str.replace(" ", "")
befolkning_1900["MeanPop"] = befolkning_1900["MeanPop"].str.replace(",", ".").astype(float)
```

```python
befolkning_1900.head()
```

```python
befolkning_1900.tail()
```

```python
befolkning_1900["Births"] = befolkning_1900["Births"].str.replace(" ", "").astype(int)
befolkning_1900
```

```python
befolkning_1900["Deaths"] = befolkning_1900["Deaths"].str.replace(" ", "").astype(int)
befolkning_1900
```

```python
sns.barplot(data=befolkning_1900, y="Year", x="Births")
```

```python
sns.lineplot(data=befolkning_1900, y="Birtrate1000", x="Year")
```
